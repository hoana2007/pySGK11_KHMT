1. NHIỆM VỤ 1

```python
def selectionSort(A):
    n = len(A)
    for i in range(n - 1):
        iMin = i
        for j in range(i + 1, n):
            if A[j] < A[iMin]:
                iMin = j
        A[i], A[iMin] = A[iMin], A[i]

infname = open("kho.inp")
soLuongDS = []
for line in infname.readlines():
    soLuongDS.append(int(line.strip()))
selectionSort(soLuongDS)
print("Số lượng các mặt hàng trong kho:")
for i in range(len(soLuongDS)):
    print(soLuongDS[i])
```

**Nhiệm vụ 2:**

```python
#Thuật toán SX chọn
def selectionSort(A):
    n = len(A)
    for i in range(n - 1):
        iMin = i
        for j in range(i + 1, n):
            if A[j] < A[iMin]:
                iMin = j
        A[i], A[iMin] = A[iMin], A[i]

#Đọc file:
infile = open("diem.inp", encoding="utf-8")
diem_ds = []
for line in infile.readlines():
    diem_ds.append(float(line))
selectionSort(diem_ds)
print("DS điểm giảm dần là: ")
for i in range(len(diem_ds)):
    print(diem_ds[i])
```

***VẬN DỤNG***

Bảng mua hàng gồm:
Tên mặt hàng, đơn giá và số lượng

```python
fname = open("muahang.inp", "r", encoding="UTF-8")
data = fname.readlines()
fname.close()
print(data)
def nhapDL(data):
    tenhang = []
    dongia = []
    soluong = []
    damua = []
    for line in data:
        L = line.split()
        tenhang = L[0]
        dongia = float(L[1])
        soluong = int(L[2])
        thanhtien = dongia * soluong
        damua.append((tenhang, thanhtien))
    return damua

def bubbleSort(A):
    n = len(A)
    for i in range(n):
        for j in range(0, n - i - 1):
            if A[j][1] > A[j + 1][1]:
                A[j], A[j+1] = A[j + 1], A[j]
mh = nhapDL(data)
bubbleSort(mh)
for i in range(len(mh)):
    print(f"{mh[i][0]} {mh[i][1]}")
```
