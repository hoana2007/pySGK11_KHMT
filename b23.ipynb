```python
# TÍNH TỔNG CÁC SỐ TỪ 1 ĐẾN N
#Cách 1: Dùng vòng lặp, cách này có độ phức tạp O(n)
def sum_of_numbers(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total
#Cách 2: Dùng công thức, cách này có độ phức tạp O(1)
def sum_of_numbers_formula(n):
    return n * (n + 1) // 2

#Cách 3: Dùng đệ quy, cách này có độ phức tạp O(n)
def sum_of_numbers_recursive(n):
    if n == 1:
        return 1
    else:
        return n + sum_of_numbers_recursive(n - 1)
    
#Nhập số n từ người dùng
n = 999
print(f"Tổng các số từ 1 đến {n} là: {sum_of_numbers_formula(n)}")
print("########################################")
print("#THỜI GIAN CHẠY CÁC CÁCH")
print("########################################")
from time import perf_counter
ALGORITHMS = {
    'Cách 1 (Vòng lặp)':    sum_of_numbers,
    'Cách 2 (Công thức)':   sum_of_numbers_formula,
    'Cách 3 (Đệ quy)':     sum_of_numbers_recursive
}

import sys
sys.setrecursionlimit(10000)  # Tăng giới hạn đệ quy nếu cần thiết

for algo_names, func in ALGORITHMS.items():
    start_time = perf_counter()
    sorted_arr = func(n)
    end_time = perf_counter()
    print(f"{algo_names:<20} : {end_time - start_time:.6f} seconds")
```

```python
#TT Sắp xếp nổi bọt
def bubbleSort(A):
    n = len(A)
    for i in range(n -1):
        for j in range(n -1-i):
            if (A[j] > A[j+1]):
                A[j], A[j+1] = A[j + 1], A[j]

#TT Quicksort
def quickSort(A):
    if len(A) < 2:
        return A
    else:
        pivot = A[0]
        left = [x for x in A[1:] if x <= pivot]
        right = [x for x in A[1:] if x > pivot]
        return quickSort(left) + [pivot] + quickSort(right)
    
#Đọc dữ liệu từ tệp vào dãy A:
file_path = "dayso.inp"
with open(file_path, "r") as f:
    A = [int(x) for x in f.read().split()]
print("Độ dài của A: ", len(A))
from time import perf_counter
ALGORITHMS = {
    'Bubble Sort': bubbleSort,
    'Quick Sort': quickSort
}
for algo_name, func in ALGORITHMS.items():
    A_copy = A.copy()  # Tạo bản sao của mảng để sắp xếp
    start_time = perf_counter()
    if algo_name == 'Quick Sort':
        sorted_A = func(A_copy)  # Quick Sort trả về mảng đã sắp xếp
    else:
        func(A_copy)  # Bubble Sort sắp xếp tại chỗ
    end_time = perf_counter()
    print(f"{algo_name:<20} : {end_time - start_time:.6f} seconds")
```

```python
# ══════════════════════════════════════════════════════════════
#  THUẬT TOÁN SẮP XẾP
# ══════════════════════════════════════════════════════════════

def bubble_sort(arr):
    a = arr[:]
    n = len(a)
    for i in range(n):
        for j in range(0, n - i - 1):
            if a[j] > a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]
    return a


def selection_sort(arr):
    a = arr[:]
    n = len(a)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if a[j] < a[min_idx]:
                min_idx = j
        a[i], a[min_idx] = a[min_idx], a[i]
    return a


def insertion_sort(arr):
    a = arr[:]
    for i in range(1, len(a)):
        key = a[i]
        j = i - 1
        while j >= 0 and a[j] > key:
            a[j + 1] = a[j]
            j -= 1
        a[j + 1] = key
    return a


def merge_sort(arr):
    if len(arr) <= 1:
        return arr[:]
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    result, i, j = [], 0, 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i]); i += 1
        else:
            result.append(right[j]); j += 1
    return result + left[i:] + right[j:]


def quick_sort(arr):
    a = arr[:]
    def _qs(a, lo, hi):
        if lo < hi:
            pivot = a[hi]
            i = lo - 1
            for j in range(lo, hi):
                if a[j] <= pivot:
                    i += 1
                    a[i], a[j] = a[j], a[i]
            a[i + 1], a[hi] = a[hi], a[i + 1]
            p = i + 1
            _qs(a, lo, p - 1)
            _qs(a, p + 1, hi)
    _qs(a, 0, len(a) - 1)
    return a


def heap_sort(arr):
    a = arr[:]
    n = len(a)
    def heapify(a, n, i):
        largest = i
        l, r = 2 * i + 1, 2 * i + 2
        if l < n and a[l] > a[largest]: largest = l
        if r < n and a[r] > a[largest]: largest = r
        if largest != i:
            a[i], a[largest] = a[largest], a[i]
            heapify(a, n, largest)
    for i in range(n // 2 - 1, -1, -1):
        heapify(a, n, i)
    for i in range(n - 1, 0, -1):
        a[0], a[i] = a[i], a[0]
        heapify(a, i, 0)
    return a


def shell_sort(arr):
    a = arr[:]
    n = len(a)
    gap = n // 2
    while gap > 0:
        for i in range(gap, n):
            temp = a[i]
            j = i
            while j >= gap and a[j - gap] > temp:
                a[j] = a[j - gap]
                j -= gap
            a[j] = temp
        gap //= 2
    return a

def sapxep(arr):
    a = arr[:]
    a.sort()
    return a

#Đọc dữ liệu từ tệp vào dãy A:
file_path = "dayso.inp"
with open(file_path, "r") as f:
    A = [int(x) for x in f.read().split()]
print("Độ dài của A: ", len(A))
print("########################################")
print("#THỜI GIAN CHẠY CÁC THUẬT TOÁN SẮP XẾP")
print("########################################")
from time import perf_counter
ALGORITHMS = {
    'Bubble Sort':    bubble_sort,
    'Selection Sort': selection_sort,
    'Insertion Sort': insertion_sort,
    'Merge Sort':     merge_sort,
    'Quick Sort':     quick_sort,
    'Heap Sort':      heap_sort,
    'Shell Sort':     shell_sort,
    'Mặc định (sort)': sapxep,
}

for algo_names, func in ALGORITHMS.items():
    start_time = perf_counter()
    sorted_arr = func(A)
    end_time = perf_counter()
    print(f"{algo_names:<20} : {end_time - start_time:.6f} seconds")
```

```python
#So sánh thời gian các thuật toán tính các số Fibonacci
#TTĐệ quy, độ phức tạp O(2^n)
def fib_DQ(n):
    return n if n <= 1 else fib_DQ(n-1) + fib_DQ(n-2)

#TTĐệ quy có ghi nhớ (memoization), độ phức tạp O(n)
def fib_memo(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = fib_memo(n-1, memo) + fib_memo(n-2, memo)
    return memo[n]

#TTTuyến tính (iterative), độ phức tạp O(n)
def fib_iter(n):        
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

#công thức, độ phức tạp O(1)
def fib_formula(n):
    phi = (1 + 5 ** 0.5) / 2
    return round((phi ** n - (1 - phi) ** n) / (5 ** 0.5))

#Đo thời gian tính của các thuật toán Fibonacci
n = 12
ALGORITHMS = {
    'Đệ quy': fib_DQ,
    'Memoization': fib_memo,
    'Tuyến tính': fib_iter,
    'Công thức': fib_formula,
}
from time import perf_counter
for algo_name, func in ALGORITHMS.items():
    start_time = perf_counter()
    result = func(n)
    end_time = perf_counter()
    print(f"{algo_name:<15} : F({n}) = {result}, thời gian: {end_time - start_time:.6f} seconds")
```
