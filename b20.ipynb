Nhiệm vụ 3: Kiểm tra điểm thi

```python
#TT tìm kiếm nhị phân
def binary_search(arr, x):
    left = 0
    right = len(arr) - 1
    while left <= right:
        mid = left + (right - left) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            left = mid + 1
        else:
            right = mid - 1
    return -1

#Đọc dữ liệu từ tệp
in_fname = open("diemthi_sx.inp")
ds_diem = []
for line in in_fname.readlines():
    ds_diem.append(float(line))
in_fname.close()
diemCanKT = float(input("Nhập điểm cần kiểm tra: "))
viTri = binary_search(ds_diem, diemCanKT)
if viTri != -1:
    print(f"Tìm thấy điểm {diemCanKT} tại vị trí {viTri}")
else:
    print(f"Không tìm thấy điểm {diemCanKT} trong danh sách")
```
