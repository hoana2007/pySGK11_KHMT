BÀI TOÁN THÁP HÀ NỘI

```python
#THUẬT TOÁN THÁP HÀ NỘI
def HaNoi(n, i, j, k):
    if n == 1:
        print("Chuyển đĩa", n, "từ cọc", i, "sang cọc", j)
    else:
        HaNoi(n - 1, i, k, j)
        print("Chuyển đĩa", n, "từ cọc", i, "sang cọc", j)
        HaNoi(n - 1, k, i, j)
#Số lần chuyển đĩa
def H(n):
    return 2**n -1
#Test:
n = 3
HaNoi(n,"A","B","C")
print("Số lần chuyển đĩa: ", H(n))

```

VẬN DỤNG:
Bài 1. CHỨNG MINH BẰNG PHƯƠNG PHÁP QUY NẠP: H(n)
1. Cơ sở quy nạp (Base Case)Xét với n = 1:
- Theo công thức: H(1) = 2^1 - 1 = 1.
- Thực tế: Để chuyển 1 đĩa từ cọc này sang cọc khác, ta chỉ cần 1 bước duy nhất.
- Vậy mệnh đề đúng với n = 1.

2. Giả thiết quy nạp (Inductive Hypothesis)
Giả sử công thức đúng với n = k (k >= 1). Nghĩa là ta có:
H(k) = 2^k - 1(Đây là số bước tối thiểu để chuyển k chiếc đĩa).

3. Bước quy nạp (Inductive Step)
Ta cần chứng minh công thức cũng đúng với n = k + 1. 
Nghĩa là chứng minh: H(k+1) = 2^(k+1) - 1
Phân tích logic: Để chuyển k+1 chiếc đĩa từ cọc A sang cọc C, quy trình thực hiện như sau:
- Bước 1: Chuyển k đĩa trên cùng từ cọc A sang cọc trung gian B. (Mất H(k) bước).
- Bước 2: Chuyển đĩa lớn nhất (đĩa thứ k+1) từ cọc A sang cọc đích C. (Mất 1 bước).
- Bước 3: Chuyển k đĩa từ cọc trung gian B về cọc đích C (đè lên đĩa lớn nhất). (Mất H(k) bước).
Thiết lập biểu thức:Tổng số bước H(k+1) sẽ là:
H(k+1) = H(k) + 1 + H(k)
H(k+1) = 2H(k) + 1
Thay giả thiết quy nạp H(k) = 2^k - 1 vào biểu thức trên:
H(k+1) = 2(2^k - 1) + 1
H(k+1) = 2.2^k - 2 + 1
H(k+1) = 2^(k+1) - 1
4. Kết luận
Vì mệnh đề đúng với n = 1 và nếu đúng với n = k thì cũng đúng với n = k + 1, nên theo nguyên lý quy nạp toán học, công thức H(n) = 2^n - 1 đúng với mọi số nguyên dương n.

```python
#Tính H(n) bằng đệ quy
def H(n):
    if n == 1:
        return 1
    else:
        return 2*H(n-1) + 1
#Test:
n = 3
print(H(n))
```

```python
#Bài 2: Lưu số bước chuyển dạng (k,j,k) vào một danh sách
def HaNoi(n, i, j, k, buocChuyen):
    if n == 1:
        buocChuyen.append((n, i, j))
        return
    else:
        HaNoi(n - 1, i, k, j, buocChuyen)
        buocChuyen.append((n, i, j))
        HaNoi(n - 1, k, i, j, buocChuyen)
def H(n):
    return 2**n - 1
#Test
n = 3
cacBuoc = []
#Số các bước
print(H(n))
HaNoi(n, 1, 3, 2, cacBuoc)
for i in range(len(cacBuoc)):
    print(cacBuoc[i], end="\n")
```
