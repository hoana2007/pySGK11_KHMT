TÌM KIẾM NHỊ PHÂN


```python
#Thuật toán
def binarySearch(A, K):
    left, right = 0, len(A) - 1
    while left <= right:
        mid = (left + right) // 2
        if A[mid] == K:
            return mid
        elif A[mid] < K:
            left = mid + 1
        else:
            right = mid - 1
    return -1
A = [0, 4, 9, 10, 12, 14, 17, 18, 20, 31, 34, 67]
K = 17
#Hiển thị số lần đã duyệt để tìm kiếm
result = binarySearch(A, K)

if result != -1:
    print("Phần tử", K, "tìm thấy tại vị trí:", result)
else:
    print("Không tìm thấy")

```

LUYỆN TẬP

1. Tìm tất cả giá trị cần tìm bằng TT Tuần tự

```python
#Thuật toán
def linearSearch(A, K):
    count = 0
    b = []
    for i in range(len(A)):
        if A[i] == K:
            b.append(i)
            count += 1
    return b,count

A = [34, 12, 56, 78, 12, 90, 11, 23]
K = 12
result, dem = linearSearch(A, K)
if result != -1:    
    for i in range(len(result)):
        print("Phần tử", K, "xuất hiện tại các vị trí:", result[i])
    print("Số lần xuất hiện: ", dem)
```

2. Tìm kiếm nhị phân với SX giảm dần

```python
#Thuật toán
def binarySearch(A, K):
    count = 0
    left, right = 0, len(A) - 1
    while left <= right:
        mid = (left + right) // 2
        count += 1
        if A[mid] == K:
            return mid, count
        elif A[mid] > K:
            left = mid + 1
        else:
            right = mid - 1
    return -1
A = [9,7,6,5,4,3,3,2,1,0]
K = 3
#Hiển thị số lần đã duyệt để tìm kiếm
result, dem = binarySearch(A, K)

if result != -1:
    print("Phần tử", K, "tìm thấy tại vị trí:", result)
    print("Số lần đã lặp để tìm được: ", dem)
```

2. Cho danh sách học sinh A. Tìm tên học sinh bằng thuật toán TK nhị phân

```python
#Thuật toán TK nhị phân
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid  # Tìm thấy, trả về vị trí
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1  # Không tìm thấy

# Danh sách lớp (đã sắp xếp)
A = ["An", "Bình", "Cường", "Minh", "Nam", "Sơn", "Trung"]
target_name = "An"

# Tìm kiếm
result = binary_search(A, target_name)

if result != -1:
    print(f"Học sinh {target_name} được tìm thấy tại vị trí: {result}")
else:
    print(f"Học sinh {target_name} không tồn tại trong danh sách.")
```
