NHIỆM VỤ 1: SẮP XẾP CHÈN BẰNG ĐỀ QUY

```python
#Thuật toán sắp xếp chèn bằng đệ quy
def insertionSortRec(A, i):
    if i > 0:
        insertionSortRec(A, i - 1)
        value = A[i]
        j = i - 1
        while j >= 0 and A[j] > value:
            A[j + 1] = A[j]
            j = j - 1
        A[j + 1] = value
#Test
A = [1,4,3,5,10,8,7,6,8]
n = len(A)
insertionSortRec(A, n - 1)
print(A)
```

NHIỆM VỤ 2: CHUYỂN SỐ THẬP PHÂN SANG NHỊ PHÂN

```python
#Thuật toán
def binary(n):
    if n == 0:
        return "0"
    elif n == 1:
        return "1"
    else:
        return binary(n//2) + str(n % 2)

#Test
n = 15
print(binary(n))
```

```python
#1. Chuyển thập phân sang nhị phân, đầu ra là 1 list
#Thuật toán
def binary(n, kq):
    if n == 0:
        kq.append("0")
    elif n == 1:
        kq.append("1")
    else:
        binary(n//2, kq)
        kq.append(str(n % 2))
    return kq

#Test
n = 10
kqDay = []
binary(n, kqDay)
print(kqDay)
```

```python
#Mở rộng - Cách viết đổi thập phân sang nhi phân "Pythonic"
def binary_steps(n):
    steps = []
    while n > 0:
        bit = n % 2
        steps.append(f"{n}/2 dư {bit}")
        n = n // 2
    return steps[::-1]

n = 25
print(binary_steps(n))
```

```python
#2. Chuyển nhị phân sang thập phân
def decimals(bin):
    if not bin:
        return 0
    # Công thức: (Bit đầu tiên * 2^(độ dài còn lại)) + Đệ quy phần còn lại
    return int(bin[0]) * (2**(len(bin) - 1)) + decimals(bin[1:])

#Test
bin = "1010"
print(decimals(bin))
```

#Bài 1: \
*Cho dãy số A, có n phần tử. Cặp A[i] và A[j] được gọi là nghịch đảo đếu j < j nhưng A[i] > A[j]. Viết chương trình đếm số cặp trong A.* \
#a) Viết chương trình không đệ quy

```python
#a) Viết chương trình không đệ quy
def soCap(A):
    dem = 0
    for i in range(len(A)):
        for j in range(i + 1, len(A)):
            if A[i] > A[j]:
                dem += 1
    return dem

#Test
A = [2, 4, 1, 3, 5]
print(soCap(A))
```

```python
#b) Viết chương trình  đệ quy
def soCapDQ(A):
    if len(A) < 2:
        return 0
    f = A[0]
    rest = A[1:]
    count_ = sum(1 for x in rest if f > x)
    return count_ + soCapDQ(rest)

#Test
A = [2, 4, 1, 3, 5]
print(soCap(A))
```

**2. Tính đa thức: F(A,x,n) = an.x^n + an-1.x^n-1 +... + a0.x^0** \
Thuật toán Horner (Tối ưu nhất)Thuật toán Horner giúp giảm số lượng phép nhân đáng kể bằng cách đặt nhân tử chung. Công thức truy hồi là:\
F = (...((a_n . x + a_{n-1}) . x + a_{n-2}) . x + ... + a_0)


```python
#Tính đa thức không đệ quy
def F(A, x):
    res = 0
    for a in reversed(A):
        res = res * x + a
    return res

#Test
A = [1,2,3]
print(F(A, 2))
```

```python
#Tính đa thức bằng đệ quy
def F(A, x, n):
    if n == len(A) - 1:
        return A[n]
    return A[n] + x * F(A, x, n + 1)

#Test
A = [1,2,3]
print(F(A, 2, 0))
```
