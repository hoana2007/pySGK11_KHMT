```python
marks = []
line = input("Nhập điểm các môn, cách nhau bởi dấu cách: ")
marks = [float(x) for x in line.split()]
total = 0
Min = marks[0]
Max = marks[0]
for m in marks:
    total += m
    if m < Min:
        Min = m
    if m > Max:
        Max = m
print("Điểm đã nhập", marks)
print("Điểm trung bình:", round(total / len(marks),2))
print("Điểm cao nhất:", Max)
print("Điểm thấp nhất:", Min)
```

Nhiệm vụ 2:

```python
names = []
marks_all = []
n = int(input("Nhập số học sinh: "))

for i in range(n):
    name = input("Nhập tên học sinh thứ: " + str(i+1) + ": ")
    names.append(name)
    line = input("Nhập điểm các môn của học sinh thứ" + str(i+1) + ", cách nhau bởi dấu cách: ")
    marks = [float(x) for x in line.split()]
    marks_all.append(marks)

#Định nghĩa các biến để thống kê
max_avg = 0
index_max_avg = 0
Min = marks_all[0][0]

#Duyệt qua mảng 2 chiều
for i in range(n):
    total = 0
    for j in range(len(marks_all[i])):
        total = total + marks_all[i][j]
        if Min > marks_all[i][j]:
            Min = marks_all[i][j]
    avg = total/len(marks_all[i])
    print("Điểm trung bình của học sinh", names[i], "là:", round(avg,2))
if max_avg < avg:
    max_avg = avg
    index_max_avg = i

#In kết quả
print(names[index_max_avg], "là học sinh có điểm trung bình cao nhất:", round(max_avg,2))
print("Điểm thấp nhất trong lớp là:", Min)
```

VẬN DỤNG

1. Đếm số lần xuất hiện tên trong mảng Tên

```python
#Đem số lần xuất hiện của các Tên trong mảng a
def lap(a):
    b, c = [],[]
    for x in a:
        if x not in b:
            b.append(x)
            c.append(1)
        else:
            k = b.index(x)
            c[k] += 1
    return b,c
a = ["Trường", "Tuấn", "Hoàng", "Dương", "Thảo", "Thảo", "Tuấn", "Lan"]
b, c = lap(a)
for i in range(len(b)):
    print("Tên", b[i], " xuất hiện", c[i], " lần")
```

VẬN DỤNG 2

```python
m = int(input("Nhập số hàng của mảng: "))
n = int(input("Nhập số cột của mảng: "))
a = [
    [1,2,3],
    [3,2,3],
    [7,3,9],
    ]
for i in range(m):
    row = []
    for j in range(n):
        val = int(input(f"Nhập phần tử a[{i}][{j}]: "))
        row.append(val)
    a.append(row)
print(a)

#a. Tính tổng các phần tử trong ma trận A:
total = 0
for i in range(m):
    for j in range(n):
        total += a[i][j]
print("Tổng các phần tử trong ma trận A:", total)

#b. In ra dòng có tổng các phần tử lớn nhất
max_row_sum = None
index_max_row = -1
for i in range(m):
    tongDong = sum(a[i])
    if max_row_sum is None or tongDong > max_row_sum:
        max_row_sum = tongDong
        index_max_row = i
print("Dòng có tổng các phần tử lớn nhất là dòng", index_max_row + 1, "với tổng là", max_row_sum)   

#c. In ra các phần tử phân biệt trong ma trận
b = []
for i in range(m):
    for j in range(n):
        if a[i][j] not in b:
            b.append(a[i][j])
print("Các phần tử phân biệt trong ma trận:", b)

#d. Tìm số lần xuất hiện số bất kì trong ma trận tại các vị trí cụ thể
x = int(input("Nhập số cần đếm số lần xuất hiện trong ma trận: "))
count = 0
positions = []
for i in range(m):
    for j in range(n):
        if a[i][j] == x:
            count += 1
            positions.append((i, j))
print(f"Số {x} xuất hiện {count} lần tại các vị trí: {positions}")
```
