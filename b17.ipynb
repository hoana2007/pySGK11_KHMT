1. Viết hàm UnitMatrix()

```python
#Viết ma trận nhị phân bậc n, đường cheo chính là 1, các phần tử còn lại là 0
def UnitMatrix(n):
    a = []
    for i in range(n):
        row = []
        for j in range(n):
            if i == j:
                row.append(1)
            else:
                row.append(0)
        a.append(row)
    return a
n = int(input("Nhập bậc ma trận n: "))
unit_matrix = UnitMatrix(n)
for i in range(n):
    for j in range(n):
        print(unit_matrix[i][j], end=' ')
    print()
```

```python
#UnitMatrix cach 2
def UnitMatrix(n):
    a = []
    for i in range(n):
        temp = [0] * i + [1] + [0] * (n - i - 1)
        a.append(temp)
    return a
n = int(input("Nhập bậc ma trận n: "))
unit_matrix = UnitMatrix(n)
for i in range(n):  
    for j in range(n):
        print(unit_matrix[i][j], end=' ')
    print()
```

2. Đếm số lần xuất hiện của số

```python
#Đem số lần xuất hiện của các số trong mảng a
def lap(a):
    b, c = [],[]
    for x in a:
        if x not in b:
            b.append(x)
            c.append(1)
        else:
            k = b.index(x)
            c[k] += 1
    return b,c
a = [0, 1, 5,7,0,2,5,5,1,1,2]
b, c = lap(a)
for i in range(len(b)):
    print("Số", b[i], " xuất hiện", c[i], " lần")
```
