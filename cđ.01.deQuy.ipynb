```python
# Viết chương trình in và đếm xuôi từ 1-100 bằng đệ quy
def in_xuoi(n):
    if n <= 100:
        print(n, end=' ')
        if (n % 10 == 0):
            print()  # Xuống dòng sau mỗi 10 số
        in_xuoi(n + 1)

in_xuoi(1)
```

```python
#2. Tính số Lucas thứ n
def lucas(n):
    if n == 0:
        return 2
    elif n == 1:
        return 1
    else:
        return lucas(n - 1) + lucas(n - 2)
n = 4
print(f"Số Lucas thứ {n} là: {lucas(n)}")

##Thử
for i in range(10):
    print(f"Số Lucas thứ {i} = ", lucas(i), end="\n")
```

```python
#Số Pell thứ n
def pell(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return 2 * pell(n - 1) + pell(n - 2)
n = 5
print(f"Số Pell thứ {n} là: {pell(n)}")

a = []
for i in range(n):
    #print(f"Số Pell thứ {i} = ", pell(i), end="\n")
    a.append(pell(i))
print(a)
```
