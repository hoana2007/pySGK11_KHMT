NHIỆM VỤ 1: BÀI TOÁN BIỂN ĐỐI SỐ NHỊ PHÂN SANG THẬP PHÂN


```python
def dec(s,n):
    if n == 1:
        return int(s[0])
    else:
        return 2*dec(s, n-1) + int(s[n-1])
#Ví dụ:
s = "11111111"
n = len(s)
print(dec(s, n))
```

NHIỆM VỤ 2: TÌM ƯỚC CHUNG 2 SỐ

```python
#Algorithm gcd(a,b)
def gcd(a,b):
    if b == 0:
        return a
    else:
        return gcd(b, a % b)
#Test:
a = 3
b = 9
print(gcd(a,b))
```

VẬN DỤNG

```python
#1. VIẾT LẠI TT TÌM UCLN BẰNG KỸ THUẬT ĐỆ QUY
def gcd(a,b):
    while a != b:
        if a < b:
            b = b - a
        else:
            a = a - b
    return a

#Dùng đệ quy
def gcdDQ(a,b):
    if b == 0:
        return a
    else:
        if a < b:
            return gcd(a, b - a)
        else:
            return gcd(a - b, b)

#Test:
a = 15
b = 9
print(gcd(a,b))
print(gcdDQ(a,b))
```

```python
#2. Viết lại TT gcd(a,b) theo thuật toán Euclid nhưng không dùng đệ quy
def gcd(a,b):
    while b != 0:
        r = a % b
        a = b
        b = r
    return a
#Test:
a = 15
b = 9
print(gcd(a,b))
```

```python
#3. Hiển thị tên và chiều cao bạn thấp nhất và cao nhất trong lớp
def min_max_height(hs):
    if not hs:
        return None, None
    min = hs[0]
    max = hs[0]
    for i in range(len(hs)):
        if min > hs[i]:
            min = hs[i]
        if max < hs[i]:
            max = hs[i]
    return min, max

#Đọc dữ liệu từ tệp
f_in = open("chieucao.inp", "r", encoding="utf-8")
tenHS = []
chieuCao = []
data = f_in.readlines()
for i in range(len(data)):
    l = data[i].split(",")
    tenHS.append(l[0])
    chieuCao.append(float(l[1]))
f_in.close()

a, b = min_max_height(chieuCao)

f_out = open("ketqua.out", "w", encoding="utf-8")
for i in range(len(chieuCao)):
    if (a == chieuCao[i]):
        print(tenHS[i]," ",chieuCao[i], file=f_out)
    if (b == chieuCao[i]):
        print(tenHS[i]," ",chieuCao[i], file=f_out)
print("MỞ TỆP ketqua.out ĐỂ XEM KẾT QUẢ")
```
