1. SẮP XẾP CHÈN

```python
#Thuật toán SX chèn
def insertionSort(A):
    n = len(A)
    for i in range(1, n):
        value = A[i]
        j = i - 1
        while j >= 0 and A[j] > value:
            A[j + 1] = A[j]
            j -= 1
        A[j + 1] = value

#Ví dụ:
A = [1, 2, 9, 5, 3, 7]
insertionSort(A)
print("Mảng đã sắp xếp:", A)
```

2. SẮP XẾP CHỌN

```python
#Thuật toán SX chọn
def selectionSort(A):
    n = len(A)
    for i in range(n - 1):
        iMin = i
        for j in range(i + 1, n):
            if A[j] < A[iMin]:
                iMin = j
        A[i], A[iMin] = A[iMin], A[i]
#Ví dụ:
A = [2,6,8,7,3,7]
selectionSort(A)
print(A)
```

3. SẮP XẾP NỔI BỌT 

```python
#TT SX nổi bọt
def bubbleSort(A):
    n = len(A)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if A[j] > A[j + 1]:
                A[j], A[j + 1] = A[j + 1], A[j]
#Ví dụ:
A = [5,3,8,6,2,7]
bubbleSort(A)
print(A)
```

1. Cho dãy A = [5, 8, 1, 0, 10, 4, 3].
**Viết CT để dãy A được sắp xếp bằng 1 trong 3 thuật toán: Chèn, Chọn và nổi bọt**


```python
#TT SX chèn
def insertionSort(A):
    n = len(A)
    for i in range(1, n):
        value = A[i]
        j = i - 1
        while j >= 0 and A[j] > value:
            A[j + 1] = A[j]
            j -= 1
        A[j + 1] = value
##TT SX Chọn
def selectionSort(A):
    n = len(A)
    for i in range(n - 1):
        iMin = i
        for j in range(i + 1, n):
            if A[j] < A[iMin]:
                iMin = j
        A[i], A[iMin] = A[iMin], A[i]
##TT SX Nổi bọt
def bubbleSort(A):
    n = len(A)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if A[j] > A[j + 1]:
                A[j], A[j + 1] = A[j + 1], A[j]

A = [5, 8, 1, 0, 10, 4, 3]
chon = input("Chọn TT cần SX: ")
while chon != "0":
    match chon:
        case "1":
            insertionSort(A)
            print("Dãy A sau SX Chèn: ", A)
        case "2":
            selectionSort(A)
            print("Dãy A sau SX Chọn: ", A)
        case "3":
            bubbleSort(A)
            print("Dãy A sau SX Nổi bọt: ", A)
        case _:
            print("Vui lòng chọn 1-3")
    chon = input("Chọn từ 1-3, chọn 0 để kết thúc: ")
```

2. Nhập 1 dãy vào từ bàn phím, các số cách nhau bởi dấu cách \n
Sắp xếp và in kết quả lên màn hình


```python
#TT Sắp xếp nổi bọt
def bubbleSort(A):
    n = len(A)
    for i in range(n -1):
        for j in range(n -1-i):
            if (A[j] > A[j+1]):
                A[j], A[j+1] = A[j + 1], A[j]

#Nhập dãy
nhapDay = input("nhap")
A = [int(x) for x in nhapDay.split()]
bubbleSort(A)
print("Dãy sau khi sắp xếp:", A)
```
