```python
#Luyện tập 
#1. Tổng các số lẻ bằng đệ quy
def tongLe(n):
    if n == 0:
        return 0
    else:
        if n % 2 != 0:
            return n + tongLe(n - 1)
        else:
            return tongLe(n - 1)
n = 10
print("Tổng các số lẻ từ 1 đến", n, "là:", tongLe(n))
```

```python
#2. In dãy A theo thứ tự ngược bằng đệ quy
def inNguoc(A, n):
    if n < 0:
        return
    else:
        print(A[n]," ", end="")
        inNguoc(A, n - 1)
A = [1,2,3,5,4,5]
inNguoc(A, len(A) - 1)
```

```python
#Vận dụng
#1. Tính tổng S = 1! + 2! ... + n!
#Không đệ quy
def giaiThua1(n):
    temp = 1
    for i in range(1, n + 1):
        temp *= i
    return temp
#Có đệ quy
def giaiThua2(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * giaiThua2(n - 1)

tong = 0
n = 10
for i in range(1, n + 1):
    tong += giaiThua2(i)
print("Tổng S = 1! + 2! + ... +", n,"! =", tong)
```

```python
#VD 2: Sắp xếp chèn bằng đề quy
def InsertionSort1(A, n):
    n = len(A)
    for i in range(1,n):
        value = A[i]
        j = i - 1
        while j >= 0 and A[j] > value:
            A[j + 1] = A[j]
            j = j - 1
        A[j + 1] = value
    
#Đệ quy
def InsertionSort2(A, n):
    if n <= 1:
        return
    InsertionSort2(A, n - 1)
    value = A[n - 1]
    j = n - 2
    while j >= 0 and A[j] > value:
        A[j + 1] = A[j]
        j = j - 1
    A[j + 1] = value

A = [5,2,4,6,8,9,10,1,3]
InsertionSort2(A, len(A))
print("Mảng sau khi sắp xếp là:", A)
```

```python
#3. TT TK nhị phân của bạn An
def binarySearch(A, left, right, K):
    if left == right:
        if A[left] == K:
            return left
        else:
            return -1
    else:
        mid = (left + right)//2
        if A[mid] == K:
            return mid
        elif A[mid] < K:
            return binarySearch(A, mid + 1, right, K)
        else:
            return binarySearch(A, left, mid, K)

#a) CT có đúng không:
A = [1, 2, 3, 4, 5, 6, 8, 9, 10]
K = 1000
result = binarySearch(A, 0, len(A) - 1, K)
if result != -1:
    print("Phần tử", K, "tồn tại trong mảng tại vị trí", result)
else:
    print("Phần tử", K, "không tồn tại trong mảng")

##Kết luận: CT đúng
#b) Phần cơ sở gồm những lệnh nào?
#Phần cơ sở gồm các lệnh trong khối lệnh if left == right:
#    if A[left] == K:
#        return left
#    else:
#        return -1
```
